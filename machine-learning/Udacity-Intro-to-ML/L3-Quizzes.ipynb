# Using SVM to caregorize fake terrain data

```python
import numpy as np
from prep_terrain_data import makeTerrainData
```

## Create toy terrain data set

```python
features_train, labels_train, features_test, labels_test = makeTerrainData()
```

## Plotting the 2D SVM

```python
import matplotlib.pyplot as plt
```

```python
def plot_the_line():
    # fit the model, don't regularize for illustration purposes
    plt.scatter(features_test[:, 0], features_test[:, 1], c=labels_test, s=30, cmap=plt.cm.Paired)

    # plot the decision function
    ax = plt.gca()
    xlim = ax.get_xlim()
    ylim = ax.get_ylim()

    # create grid to evaluate model
    xx = np.linspace(xlim[0], xlim[1], 30)
    yy = np.linspace(ylim[0], ylim[1], 30)
    YY, XX = np.meshgrid(yy, xx)
    xy = np.vstack([XX.ravel(), YY.ravel()]).T
    Z = clf.decision_function(xy).reshape(XX.shape)

    # plot decision boundary and margins
    ax.contour(XX, YY, Z, colors='k', levels=[0], alpha=0.5,
               linestyles=['--', '-', '--'])
    # plot support vectors
    # ax.scatter(clf.support_vectors_[:, 0], clf.support_vectors_[:, 1], s=100,
    #            linewidth=1, facecolors='none', edgecolors='k')
    plt.show()
```

## Train SVC Model

```python
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score
```

### Quiz: Coding up the SVM

```python
clf = SVC(kernel='linear')
clf.fit(features_train, labels_train)
labels_pred = clf.predict(features_test)
print("accuracy:", accuracy_score(labels_pred, labels_test))
plot_the_line()
```

Answer: Accuracy = 0.92

### Quiz: Kernel and Gamma

```python
clf = SVC(kernel='linear', gamma=1.0)
clf.fit(features_train, labels_train)
labels_pred = clf.predict(features_test)
print("accuracy:", accuracy_score(labels_pred, labels_test))
plot_the_line()
```

```python
clf = SVC(kernel='linear', gamma=1000.0)
clf.fit(features_train, labels_train)
labels_pred = clf.predict(features_test)
print("accuracy:", accuracy_score(labels_pred, labels_test))
plot_the_line()
```

Answer: gamma doesn't seem to effect the linear kernel

### Quiz: SVM C Parameter

```python
clf = SVC(C=1000.0)
clf.fit(features_train, labels_train)
labels_pred = clf.predict(features_test)
print("accuracy:", accuracy_score(labels_pred, labels_test))
plot_the_line()
```

```python
clf = SVC(C=1.0)
clf.fit(features_train, labels_train)
labels_pred = clf.predict(features_test)
print("accuracy:", accuracy_score(labels_pred, labels_test))
plot_the_line()
```
