# Quizzes for L4, the decision tree unit

```python
import numpy as np
import matplotlib.pyplot as plt
import math

from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score

from prep_terrain_data import makeTerrainData
from class_vis import prettyPicture
```

```python
features_train, labels_train, features_test, labels_test = makeTerrainData()
```

### Quiz: Coding A Decision Tree

```python
def fit_and_display_decision_tree():
    clf = DecisionTreeClassifier()

    clf.fit(features_train, labels_train)

    prettyPicture(clf, features_test, labels_test)
fit_and_display_decision_tree()
```

### Quiz: Decision Tree Accuracy

```python
def fit_and_display_decision_tree():
    clf = DecisionTreeClassifier()

    clf.fit(features_train, labels_train)
    accuracy = accuracy_score(clf.predict(features_test), labels_test)
    print(f'accuracy of model: {accuracy}')

fit_and_display_decision_tree()
```

### Quiz: Min Samples Split

```python
def fit_and_display_decision_tree():
    clf = DecisionTreeClassifier(min_samples_split=50)

    clf.fit(features_train, labels_train)
    
    accuracy = accuracy_score(clf.predict(features_test), labels_test)
    print(f'accuracy of model: {accuracy}')

    prettyPicture(clf, features_test, labels_test)
fit_and_display_decision_tree()
```

### Quiz: Entropy Calculation

```python
def entropy(probabilities):
    return -sum(map(lambda x: x * math.log2(x), probabilities))
entropy([0.5,0.5])
```
